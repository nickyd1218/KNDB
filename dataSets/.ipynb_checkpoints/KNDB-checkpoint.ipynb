# Creating and Cleaning Data Set

```python
import pandas as pd
import csv

mvpdataload= "MVPlayerOTY_fixed.xlsx"
seasonstatload = "Seasons_Stats_fixed.csv"

sdata = pd.read_csv(seasonstatload)
mdata = pd.read_excel(mvpdataload)
```

```python
kndb = pd.merge(sdata, mdata, how='left', on=["Year","Player"])
```

```python
kndb = kndb[kndb.Year >= 1980]
```

```python
kndb_test = kndb[kndb.Year != 2017]
kndb_test = kndb_test.drop(columns = ['Player','Tm'])
```

# Linear Model Logistic Regression Machine Learning

```python
X = kndb_test.drop("Award", axis=1)
y = kndb_test["Award"]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
kndb_test[:, 2] = labelencoder_X.fit_transform(X[:,2])
onehotencoder = OneHotEncoder(categorical_features = [3])
X = onehotencoder.fit_transform(X).toarray()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
kndb_predict = kndb[kndb.Year = 2017]
predictions = classifier.predict(kndb_predict)
```
